```python
import csv
import numpy as np
import matplotlib.pyplot as plt
def read_csv(filename):
    arrivals = []
    states = []
    with open(filename, 'r') as csvfile:
        csvreader = csv.reader(csvfile)
        for row in csvreader:
            try:
                time, state = map(float, row)
                arrivals.append(time)
                states.append(int(state))
            except ValueError:
                print(f"Issue with row: {row}. Skipping this row.")
    return arrivals, states
# Example usage
filename = 'Test 1a.csv' 
arrivals, states = read_csv(filename)

# Plotting the MMPP from the CSV file
plt.figure(figsize=(10, 6))
plt.step(arrivals, states, where='post')
plt.xlabel('Time')
plt.ylabel('State')
plt.title('Markov Modulated Poisson Process from CSV')
plt.show()
```
